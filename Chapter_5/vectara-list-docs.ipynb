```python
import requests
import os,json
from dotenv import load_dotenv
load_dotenv()

CORPUS_KEY = "RAGBOOK"
VECTARA_API_KEY = os.getenv("VECTARA_API_KEY")
```

```python
url = f"https://api.vectara.io/v2/corpora/{CORPUS_KEY}/documents"
headers = {
  'Content-Type': 'application/json',
  'Accept': 'application/json',
  'x-api-key': VECTARA_API_KEY
}

response = requests.get(url, headers=headers)
res = response.json()
res
```

### **Retrieve and Summarize documents**

```python
url = f"https://api.vectara.io/v2/corpora/{CORPUS_KEY}/documents/selected-works-of-shakespeare"
headers = {
  'Content-Type': 'application/json',
  'Accept': 'application/json',
  'x-api-key': VECTARA_API_KEY
}

response = requests.get(url, headers=headers)
res = response.json()
res
```

```python
url = f"https://api.vectara.io/v2/corpora/{CORPUS_KEY}/documents/selected-works-of-shakespeare/summarize"
headers = {
  'Content-Type': 'application/json',
  'Accept': 'application/json',
  'x-api-key': VECTARA_API_KEY
}

response = requests.post(url, headers=headers, json={'llm_name': 'gpt-4o'})
res = response.json()
print(res['summary'])
```
