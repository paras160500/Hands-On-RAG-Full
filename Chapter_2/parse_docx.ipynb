```python
import os 
from docx import Document
from IPython.display import display,Image 

document = Document("Sample/sample_data.docx")

# Extract Text
for para in document.paragraphs:
    if len(para.text.strip()) > 0:               # Only print non-empty paras
        print(para.text , end = "\n-----\n")
    
# Extract Tables
for table in document.tables:
    print("\n-------Table------\n")
    for row in table.rows:
        row_text = [cell.text for cell in row.cells]
        print(row_text)
```

## **ImageExtraction**

```python
import zipfile 

zipf = zipfile.ZipFile('Sample/sample_data.docx')
filelist = zipf.namelist()
print(filelist)
```

```python
for fname in filelist:
    _ , ext = os.path.splitext(fname)
    if ext in ['.jpg','.jpeg' , '.png' , '.gif']:
        # read image and display in Jupyter
        with zipf.open(fname) as img_file:
            img_data = img_file.read()
            display(Image(data = img_data))
```
